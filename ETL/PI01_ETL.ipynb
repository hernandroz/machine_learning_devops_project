```python
import pandas as pd
import ast

df2 = pd.read_csv('/content/drive/MyDrive/datasets_colab/credits.csv')
df2.head()

#-----------------------------------------------
#Desanidaremos la columna "cast"
#-----------------------------------------------

# Reemplazamos NaN por una lista compuesta por un diccionario vacío
df2['cast'] = df2['cast'].apply(lambda x: [] if pd.isna(x) else x)

# Utilizamos una función personalizada que verifique si el valor es una cadena y
# luego intente convertirla en una lista para finalmente extraer los nombres del cast

def extract_cast_names(value):
    if isinstance(value, str):
        try:
            cast_list = ast.literal_eval(value)
            return [d.get('name') for d in cast_list]
        except (ValueError, SyntaxError):
            return None
    elif isinstance(value, list):
        return [d.get('name') for d in value]
    else:
        return None

#Creamos la columna "cast_names" compuesta por los valores de la clave
#"name" de los diccionarios de las listas de la columna "cast"

df2['cast_names'] = df2['cast'].apply(extract_cast_names)

#Reemplazamos las listas vacías por None
df2['cast_names'] = df2['cast_names'].replace([], None)

#Eliminamos la columna "cast" original
df2 = df2.drop('cast', axis=1)


#-----------------------------------------------
#Desanidaremos la columna "crew"
#-----------------------------------------------


# Reemplazamos NaN por una lista compuesta por un diccionario vacío
df2['crew'] = df2['crew'].apply(lambda x: [] if pd.isna(x) else x)

# Utilizamos una función personalizada que verifique si el valor es una cadena y
#luego intente convertirla en una lista para finalmente extraer los nombres del crew

def extract_crew_names(value):
    if isinstance(value, str):
        try:
            crew_list = ast.literal_eval(value)
            return [d.get('name') for d in crew_list]
        except (ValueError, SyntaxError):
            return None
    elif isinstance(value, list):
        return [d.get('name') for d in value]
    else:
        return None

#Creamos la columna "crew_names" compuesta por los valores de la clave
#"name" de los diccionarios de las listas de la columna "crew"

df2['crew_names'] = df2['crew'].apply(extract_crew_names)

#Reemplazamos las listas vacías por None
df2['crew_names'] = df2['crew_names'].replace([], None)

#Eliminamos la columna "crew" original
df2 = df2.drop('crew', axis=1)
df2.head()
```

**2DA CONSIGNA:** Ahora respecto al dataset "movies_dataset", algunos campos "como belongs_to_collection", "production_companies" y otros (ver diccionario de datos) están anidados, esto es o bien tienen un diccionario o una lista como valores en cada fila. Deberán desanidarse y unirlos al dataset de nuevo

```python
#Creamos el dataframe en base al dataset "movies_dataset.csv"
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/movies_dataset.csv', low_memory = False)
pd.set_option('display.max_columns', None)

#-----------------------------------------------
#Desanidaremos la columna "belongs_to_collection"
#-----------------------------------------------

#Usamos la librería "ast" para que el modelo reconozca los elementos de la columna
#"belongs_to_collection" como diccionarios y retornamos NaN en caso los valores sean NaN

import ast
import numpy as np
def safe_literal_eval(x):
    try:
        return ast.literal_eval(x)
    except (ValueError, SyntaxError, TypeError):
        return np.nan
df['belongs_to_collection'] = df['belongs_to_collection'].apply(safe_literal_eval)

#Nos quedamos solo con los valores de la clave "name" y retornamos "None" en caso
#trate de un valor NaN

belongs_to_collection_dictname = df['belongs_to_collection'].apply(lambda x: x['name'] if isinstance(x, dict) else None)

#Eliminamos la columna "belongs_to_collection" original
df = df.drop('belongs_to_collection', axis=1)

#Le agregamos a nuestro dataframe una nueva columna "collection_name" compuesta por
#los valores de la clave "name" de los diccionarios de la columna "belongs_to_collection"
df = df.assign(collection_name = belongs_to_collection_dictname)

#-----------------------------------------------
#Desanidaremos la columna "genres"
#-----------------------------------------------

import pandas as pd

# Reemplazar NaN por una lista compuesta por un diccionario vacío
df['genres'] = df['genres'].apply(lambda x: [] if pd.isna(x) else x)

# Utilizamos una función personalizada que verifique si el valor es una cadena y
#luego intente convertirla en una lista para finalmente extraer los nombres de los géneros

def extract_genre_names(value):
    if isinstance(value, str):
        try:
            genres_list = ast.literal_eval(value)
            return [d.get('name') for d in genres_list]
        except (ValueError, SyntaxError):
            return None
    elif isinstance(value, list):
        return [d.get('name') for d in value]
    else:
        return None

#Creamos la columna "genre_names" compuesta por los valores de la clave
#"name" de los diccionarios de las listas de la columna "genres"

df['genre_names'] = df['genres'].apply(extract_genre_names)

#Reemplazamos las listas vacías por None
df['genre_names'] = df['genre_names'].replace([], None)

#Eliminamos la columna "genres" original
df = df.drop('genres', axis=1)

#-----------------------------------------------
#Desanidando "production_companies"
#-----------------------------------------------

import ast
#Creamos una lista vacía "production_companies_names"con el objetivo de llenarla
#con todos los valores de la clave "name" de los diccionarios de la columna
#"production_companies". Al final deberemos convertir la lista obtenida en una
#columna de nuestro dataframe

production_companies_names=[]

for l in df['production_companies']:

#El método ast.literal_eval() hará que el sistema reconozca el tipo
#al que corresponde cada dato de la columna "production_companies", pero se
#le complicará hacerlo si el tipo de dato original del objeto no es str, por
#ello estableceremos una condicional antes

  if type(l)==str:
    l=ast.literal_eval(l)

#Explorando el dataframe se ha encontrado datos de tipo str que con el método
#ast.literal_eval() el sistema reconoce que son booleanos o de otro tipo, así
#que de encontrarse uno de estos, se les asignará el nuevo valor de una lista
#vacía

    if type(l)==list:
      if l!=[]:
        names_list=[]

        for d in l:
          names_list.append(d['name'])
        production_companies_names.append(names_list)
      else:
        production_companies_names.append(l)
    else:
      l=[]
      production_companies_names.append(l)
  else:
    l=[]
    production_companies_names.append(l)
df['production_companies_names']=production_companies_names

#Eliminamos la columna "production_companies" original

df = df.drop('production_companies',axis=1)

#-----------------------------------------------
#Desanidando "production_countries"
#-----------------------------------------------

import ast

production_countries_names=[]

for l in df['production_countries']:

  if type(l)==str:
    l=ast.literal_eval(l)

    if type(l)==list:
      if l!=[]:
        names_list=[]

        for d in l:
          names_list.append(d['name'])
        production_countries_names.append(names_list)
      else:
        production_countries_names.append(l)
    else:
      l=[]
      production_countries_names.append(l)
  else:
    l=[]
    production_countries_names.append(l)
df['production_countries_names']=production_countries_names

#Eliminamos la columna "production_countries" original

df = df.drop('production_countries',axis=1)

#-----------------------------------------------
#Desanidando "spoken_languages"
#-----------------------------------------------

import ast

spoken_languages_names=[]

for l in df['spoken_languages']:

  if type(l)==str:
    l=ast.literal_eval(l)

    if type(l)==list:
      if l!=[]:
        names_list=[]

        for d in l:
          names_list.append(d['name'])
        spoken_languages_names.append(names_list)
      else:
        spoken_languages_names.append(l)
    else:
      l=[]
      spoken_languages_names.append(l)
  else:
    l=[]
    spoken_languages_names.append(l)

df['spoken_languages_names']=spoken_languages_names

#Eliminamos la columna "spoken_languages" original

df = df.drop('spoken_languages',axis=1)

df.head()
```

**3RA CONSIGNA:** Los valores nulos de los campos 'revenue' y 'budget' deben ser rellenados por el número 0.

```python
#Usando df.isnull().sum() se comprueba que la columna "budget" no contiene valores
#nulos, mientras que la columna "revenue" posee 6 valores nulos

#Rellenamos los valores nulos de la columna "budget" por el número 0

df["revenue"]=df["revenue"].fillna(0)

#Comprobamos que la operación se realizó con éxito:
print('La cantidad de valores nulos es:')
df.isnull().sum()
```

**4TA CONSIGNA:** Los valores nulos del campo 'release_date' deben eliminarse.

```python
# Eliminamos las filas con valores nulos en la columna "release_date"
df = df.dropna(subset=["release_date"])

#Comprobamos que la columna "release_date" ya no posea valores nulos:

df.isnull().sum()

#Eliminar filas puede ocasionar la ausencia de índices en el dataframe (lo cual
#se puede comprobar usando 'print(df.index)', por ello, asignaremos índices
#correctos y eliminaremos la columna antigua de índices

df = df.reset_index(drop=True)

#Comprobamos que la operación se realizó con éxito:
print('La cantidad de valores nulos en la columna "release_date" es:')
df['release_date'].isnull().sum()
```

**5TA CONSIGNA:** De haber fechas, deberán tener el formato AAAA-mm-dd. Además, se deberá extraer el año de la fecha de estreno para crear la columna "release_year"

```python
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d', errors='coerce')

# Utilizamos una función personalizada para extraer los años de 'release_date'
def extract_release_year(value):
  try:
    anio = int(value.year)
    return int(anio)
  except (ValueError, SyntaxError):
    return None
df['release_year'] = df['release_date'].apply(extract_release_year)
df.head()
# Nota: Parece que al final los números no se pudieron transformar a formato entero,
# debido a que los demás datos de la columna son tipo float (nan) y no puede haber
# más de un tipo de datos en una columna
```

**6TA CONSIGNA:** Crear una columna compuesta por el retorno de inversión (llamada "return") y usando los campos "revenue" y "budget", dividiendo estas dos últimas revenue / budget. Cuando no haya datos disponibles para calcularlo, deberá tomar el valor 0.

```python
#Se observó que los datos numéricos de la columna 'revenue' están en formato float,
#pero los de la columna  'budget' están en formato str, así que los convertiremos
#en formato float para poder hacer la división correctamente

#Crearemos una lista vacía que mediante iteraciones terminará albergando todos
#los datos de la columna "budget" pero en formato float

budget_float=[]

for b in df['budget']:

#Se detectó que existen tres datos que no son números, así que los convertiremos
#en el número 0 y con el tipo float

  if b=='/ff9qCepilowshEtG2GYWwzt2bs4.jpg':
    b=float(0)
    budget_float.append(b)

  elif b=='/zV8bHuSL6WXoD6FWogP9j4x80bL.jpg':
    b=float(0)
    budget_float.append(b)

  elif b=='/zaSf5OG7V8X8gqFvly88zDdRm46.jpg':
    b=float(0)
    budget_float.append(b)
  else:
    b=float(b)
    budget_float.append(b)

#Ahora hacemos que a la columna "budget" le corresponda los datos flotantes obtenidos

df['budget']=budget_float

#Crearemos una lista vacía que mediante iteraciones terminará albergando todos
#los datos obtenidos de dividir revenue/budget

return_list=[]

for i in range(df.shape[0]):

# Asignaremos el número 0 si la división no es posible:

  if (df['budget'][i])==float(0):
    ret=0

    return_list.append(ret)
  else:
    ret = (df['revenue'][i]) / (df['budget'][i])
    return_list.append(ret)
df['return']=return_list
```
